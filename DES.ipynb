```python
import numpy as np
def __int_to_BitList(data):
	"""Turn the string data, into a list of bits (1, 0)'s"""
		
	l = len(data) * 8
	result = [0] * l
	pos = 0
	for ch in data:
		i = 7
		while i >= 0:
			if ch & (1 << i) != 0:
				result[pos] = 1
			else:
				result[pos] = 0
			pos += 1
			i -= 1

	return result
a=[1,2,3,4]
blist=__int_to_BitList(a)
```

```python
def __BitList_to_int(data):
	"""Turn the list of bits -> data, into a np.array([...],dtype=np.int8)"""
	result = []
	pos = 0
	c = 0
	while pos < len(data):
		c += data[pos] << (7 - (pos % 8))
		if (pos % 8) == 7:
			result.append(c)
			c = 0
		pos += 1

		
	return np.array(result,dtype=np.uint8)
	
b=__BitList_to_int(blist)
```
